```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Arcturus PEPSI 6000-6100.csv', sep=';')

fila_0 = df.iloc[0].astype(str).str.strip()
fila_0 = fila_0.replace('nan', '')

if fila_0.str.fullmatch(r'[:.,;\s]*').all():
    df = df.drop(index=0).reset_index(drop=True)

wavelength = df['Wavelength'].values

normalized_flux = df['Normalized Flux'].values
```

Revisión de línea espectral de 6012.2099 y 6012.4377

```python
def graficar_rango(df, lam_min, lam_max):
    sub_df = df[(df["Wavelength"] >= lam_min) & (df["Wavelength"] <= lam_max)]
    
    plt.figure(figsize=(8, 4))
    plt.plot(sub_df["Wavelength"], sub_df["Normalized Flux"], color="lightcoral")
    plt.xlabel("Longitud de onda [Å]")
    plt.ylabel("Flujo normalizado")
    plt.title(f"Espectro entre {lam_min}–{lam_max} Å")
    plt.grid(True)
    plt.show()
graficar_rango(df, 6010, 6013)
```
